```
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sn
import h5py
```

**Load loss**

```
def load_loss():
    filename = 'mlp_v3/loss.hdf5'

    with h5py.File(filename, 'r') as f:
        return f['data'][:]
```

**Visualizing Loss**

```
loss = load_loss()
```

```
loss
```

```
plt.figure(figsize=(20,5))
plt.plot(loss)
plt.xlabel('epoch')
plt.ylabel('loss')

plt.show()
```

**Confusion Matrix**

```
cm = [[46763, 6999], [7105, 17642]]
```

```
np_cm = np.array(cm)
```

```
plt.figure(figsize=(14,10))
ax = plt.axes()
sn.heatmap(np_cm, annot=True, fmt='d', cmap='Blues')
ax.set_xlabel('Predict label')
ax.set_ylabel('True label')

plt.show()
```
